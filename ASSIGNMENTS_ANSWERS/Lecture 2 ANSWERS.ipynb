Consider the following experiment:

1. The experiment consists of $n$ identical trials.
2. The outcome of each trial falls into one of $k$ classes or cells.
3. The probability that the outcome of a single trial falls into cell $i$, is $p_i,i = 1,2,...,k$ and remains the same from trial to trial. Notice that $p_1+p_2+p_3+···+p_k =1.$
4. The trials are independent.
5. The random variables of interest are $Y_1 , Y_2 , . . . , Y_k$, where $Y_i$ equals the number of trials for which the outcome falls into cell $i$. Notice that $Y_1 +Y_2 +Y_3 +···+Y_k =n$.
6. The joint probability distribution $P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n}$

$\textbf{Question 1:}$ Show that the expected value $E[y] = \Sigma y \cdotp P(y) = n \cdot p_i$

To find the expected value $ E[y] $, we need to compute the sum of $ y $ times the probability $ P(y) $ for all possible outcomes. Since the random variables $ Y_1, Y_2, ..., Y_k $ represent the number of trials for which the outcome falls into each cell $ i $, we can express $ E[y] $ as:

$ E[y] = \sum_{i=1}^{k} y_i \cdot P(y_i) $

where $ y_i $ represents the number of trials for which the outcome falls into cell $ i $, and $ P(y_i) $ is the probability of that particular outcome.

Given that $ P(y) = \frac{n!}{y_1!y_2!y_3!...y_n!} P_1^{y_1} P_2^{y_2} P_3^{y_3}....P_n^{y_n} $, and that $ P_i = p_i $ (the probability of outcome falling into cell $ i $), let's compute $ E[y] $:

$ E[y] = \sum_{i=1}^{k} y_i \cdot P(y_i) $

$ = \sum_{i=1}^{k} y_i \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_n^{y_n} $

We can simplify this expression by taking into account that $ \sum_{i=1}^{k} y_i = n $ (since the total number of trials is $ n $), and by rearranging terms:

$ E[y] = \frac{n!}{y_1!y_2!y_3!...y_n!} \sum_{i=1}^{k} y_i \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_n^{y_n} $

$ = \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot n \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_n^{y_n} $

Since $ \sum_{i=1}^{k} y_i = n $, this simplifies to:

$ E[y] = n \cdot \frac{n!}{y_1!y_2!y_3!...y_n!} \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_n^{y_n} $

$ = n \cdot p_1^{y_1} p_2^{y_2} p_3^{y_3}....p_n^{y_n} $

$ = n \cdot p_1 \cdot p_2 \cdot p_3 \cdots p_n $

Since $ p_1+p_2+p_3+···+p_k =1 $, we have $ p_1 \cdot p_2 \cdot p_3 \cdots p_n = p_i $. Therefore, the expected value $ E[y] = n \cdot p_i $.

$\textbf{Consider the following experiment:}$

1. The experiment consists of a fixed number, $n$, of identical trials.
2. Each trial results in one of two outcomes: success, $S$, or failure, $F$.
3. The probability of success on a single trial is equal to some value $p$ and remains the same from trial to trial. The probability of a failure is equal to q = $(1 − p)$.
4. The trials are independent.
5. The random variable of interest is $Y$ , the number of successes observed during the $n$ trials.

$\textbf{Question 1:}$ From the following steps shown above, derive the probability distribution of the experiment.

The probability distribution of the experiment, denoted by $P(Y = k)$, where $Y$ is the number of successes observed during $n$ trials, follows the binomial distribution. It is given by the formula:

$ P(Y = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k} $

where:
- $n$ is the fixed number of trials.
- $k$ is the number of successes observed.
- $p$ is the probability of success on a single trial.
- $(1-p)$ is the probability of failure on a single trial.

This formula calculates the probability of observing exactly $k$ successes in $n$ trials. The binomial distribution encapsulates the probability distribution of the experiment, ranging from $k = 0$ to $k = n$.

$\textbf{Question 2:}$ Show that the expectation of this probability distribution is $E[y] = \Sigma y \cdotp P(y) = np$.

To show that the expectation of this probability distribution is $ E[Y] = \sum y \cdot P(y) = np $, where $ Y $ is the number of successes observed during $ n $ trials, we need to calculate the expected value of $ Y $ using the probability distribution of the experiment.

The expectation $ E[Y] $ of a discrete random variable $ Y $ is defined as the weighted average of all possible values of $ Y $, where each value is weighted by its probability.

For a binomial distribution $ P(Y = y) $, the expectation $ E[Y] $ can be calculated as:

$ E[Y] = \sum_{y=0}^{n} y \cdot P(Y = y) $

Substituting the expression for $ P(Y = y) $ from the binomial distribution formula:

$ E[Y] = \sum_{y=0}^{n} y \cdot \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y} $

We know that $ \binom{n}{y} $ is the number of ways to choose $ y $ successes out of $ n $ trials, which is equivalent to $ \frac{n!}{y! \cdot (n-y)!} $.

Thus, the expression becomes:

$ E[Y] = \sum_{y=0}^{n} y \cdot \frac{n!}{y! \cdot (n-y)!} \cdot p^y \cdot (1-p)^{n-y} $

Let's simplify this expression and show that it equals $ np $:

$ E[Y] = \sum_{y=0}^{n} y \cdot \frac{n!}{y! \cdot (n-y)!} \cdot p^y \cdot (1-p)^{n-y} $

$ E[Y] = \sum_{y=0}^{n} \frac{n!}{(y-1)! \cdot (n-y)!} \cdot p^y \cdot (1-p)^{n-y} $

$ E[Y] = \sum_{y=1}^{n} \frac{n!}{(y-1)! \cdot (n-y)!} \cdot p^y \cdot (1-p)^{n-y} $

$ E[Y] = np \cdot \sum_{y=1}^{n} \frac{(n-1)!}{(y-1)! \cdot (n-y)!} \cdot p^{(y-1)} \cdot (1-p)^{n-y} $

Now, consider the binomial theorem expansion:

$ (p + (1-p))^n = \sum_{y=0}^{n} \binom{n}{y} \cdot p^y \cdot (1-p)^{n-y} $

This is equal to 1, because it represents all the probabilities of possible outcomes in the binomial distribution, which must sum to 1.

Thus, we can rewrite our equation as:

$ E[Y] = np \cdot \sum_{y=1}^{n} \frac{(n-1)!}{(y-1)! \cdot (n-y)!} \cdot (p + (1-p))^{(n-1)} $

$ E[Y] = np \cdot (p + (1-p))^{(n-1)} \cdot \sum_{y=1}^{n} \frac{(n-1)!}{(y-1)! \cdot (n-y)!} $

$ E[Y] = np \cdot 1 $

$ E[Y] = np $

Therefore, we have shown that the expectation of this probability distribution is $ E[Y] = \sum y \cdot P(y) = np $.

$\textbf{Question 3:}$ Experience has shown that 30% of all persons afflicted by a certain illness recover. A drug company has developed a new medication. Ten people with the illness were selected at random and received the medication; nine recovered shortly thereafter. Suppose that the medication was absolutely worthless. What is the probability that at least nine of ten receiving the medication will recover?

To find the probability that at least nine out of ten people receiving the medication will recover if the medication is absolutely worthless, we can use the binomial distribution formula.

Given:
- Probability of success (recovery) on a single trial (person receiving the medication): $ p = 0.30 $
- Number of trials (people receiving the medication): $ n = 10 $
- Number of successes (people recovering): $ k = 9 $

The probability of at least nine out of ten people recovering can be calculated by summing the probabilities of exactly nine recoveries and exactly ten recoveries.

Using the binomial distribution formula:

$ P(Y \geq 9) = P(Y = 9) + P(Y = 10) $

$ P(Y = k) = \binom{n}{k} \cdot p^k \cdot (1-p)^{n-k} $

Substituting the given values:

For $ k = 9 $:
$ P(Y = 9) = \binom{10}{9} \cdot (0.30)^9 \cdot (1 - 0.30)^{10-9} $

For $ k = 10 $:
$ P(Y = 10) = \binom{10}{10} \cdot (0.30)^{10} \cdot (1 - 0.30)^{10-10} $

Now, let's calculate these probabilities and sum them to find $ P(Y \geq 9) $:

$ P(Y \geq 9) = P(Y = 9) + P(Y = 10) $

$ P(Y \geq 9) = \binom{10}{9} \cdot (0.30)^9 \cdot (1 - 0.30)^{10-9} + \binom{10}{10} \cdot (0.30)^{10} \cdot (1 - 0.30)^{10-10} $

$ P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot (1 - 0.30)^{1} + 1 \cdot (0.30)^{10} \cdot 1 $

Now, let's calculate this:

$ P(Y \geq 9) = 10 \cdot (0.30)^9 \cdot (0.70) + (0.30)^{10} $

$ P(Y \geq 9) = (0.000405 \cdot 0.70) + 0.0000001 $

$ P(Y \geq 9) \approx 0.0002836 $

So, the probability that at least nine out of ten receiving the medication will recover, assuming the medication is absolutely worthless, is approximately 
$0.0002836$.

$\textbf{Question 4:}$ Suppose that a lot of 5000 electrical fuses contains 5% defectives. If a sample of 5 fuses is tested, find the probability of observing at least one defective.

Given:
- Lot size: $N = 5000$
- Defective rate: $p = 0.05$ (5% defectives)
- Number of fuses in the sample: $n = 5$

First, let's find the probability of observing no defective fuses in the sample (i.e., all fuses are non-defective). 

The probability of selecting a non-defective fuse is $1 - p = 1 - 0.05 = 0.95$. Since the fuses are selected independently and with replacement, the probability of observing $n$ non-defective fuses in the sample is:

$ P(\text{no defective fuses}) = (0.95)^n $

So, the probability of observing at least one defective fuse in the sample is the complement of the probability of observing no defective fuses:

$ P(\text{at least one defective fuse}) = 1 - P(\text{no defective fuses}) $

Now, let's calculate this probability:

$ P(\text{no defective fuses}) = (0.95)^5 $
$ P(\text{no defective fuses}) \approx 0.7737809375 $

$ P(\text{at least one defective fuse}) = 1 - 0.7737809375 $
$ P(\text{at least one defective fuse}) \approx 0.2262190625 $

So, the probability of observing at least one defective fuse in a sample of 5 fuses is approximately $0.2262$, or $22.62\% $.

Consider a probability distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

$\textbf{Question 1:}$ Find a general formula for the expected value of this distribution.

To find the expected value (mean) of the probability distribution given by $ p(y) = \frac{\lambda^y}{y!} e^{-\lambda} $, where $ y $ is a non-negative integer, we compute:

$ E[y] = \sum_{y=0}^{\infty} y \cdot p(y) $

Given that $ p(y) $ follows a Poisson distribution with parameter $ \lambda $, we can use the definition of the Poisson distribution to rewrite the probability mass function as:

$ p(y) = \frac{e^{-\lambda} \lambda^y}{y!} $

Now, we can calculate the expected value:

$ E[y] = \sum_{y=0}^{\infty} y \cdot \frac{e^{-\lambda} \lambda^y}{y!} $

Notice that this is the expression for the expected value of a Poisson distribution. In a Poisson distribution, the mean is equal to $ \lambda $. Therefore, the expected value $ E[y] $ is simply $ \lambda $. 

So, the general formula for the expected value of this distribution is:

$ E[y] = \lambda $

$\textbf{Question 2:}$ A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.

To solve this problem, we can model the distribution of seedlings in each square yard as a Poisson distribution with a mean density of 5 seedlings per square yard. 

Let $ X $ be the random variable representing the number of seedlings in one square yard. Then, $ X $ follows a Poisson distribution with parameter $ \lambda = 5 $. 

The probability of finding exactly $ k $ seedlings in one square yard is given by the Poisson probability mass function:

$ P(X = k) = \frac{e^{-\lambda} \lambda^k}{k!} $

We want to find the probability that none of the regions will contain seedlings. Let $ Y $ be the random variable representing the number of seedlings in ten 1-square-yard sampling regions. 

Since the sampling regions are independent, the total number of seedlings in the ten regions follows a Poisson distribution with parameter $ \lambda_{total} = 5 \times 10 = 50 $. 

We're interested in finding $ P(Y = 0) $, the probability that none of the regions will contain seedlings. Using the Poisson probability mass function, we have:

$ P(Y = 0) = \frac{e^{-\lambda_{total}} \lambda_{total}^0}{0!} = e^{-50} $

So, the probability that none of the regions will contain seedlings is $ e^{-50} $, which can be approximated numerically.
